# Análise exploratória Covid19

Para análise iremos utilizar dois arquivos CSV  disponiblizados na plataforma Kaggle através do link: [DataSet](https://www.kaggle.com/imdevskp/corona-virus-report?select=worldometer_data.csv)

- **full_grouped.csv**: Contém dados separados por dia desde 2019-01-22 até 2020-07-27.

- **worldometer_data.csv**: Medidor Mundial com dados separados por país/região

```
medidorMundo = pd.read_csv('worldometer_data.csv')
medidorMundo.head()
```

#### Avaliando DataFrame

```
medidorMundo.info()
```

```
medidorMundo.isnull().sum()
```

#### Dropando valores faltantes e efetuando agrupamento por regiões. 

```
medidorMundo_tabela = medidorMundo.drop(['NewCases','NewDeaths','NewRecovered'],axis=1)
medidorMundo_tabela = medidorMundo_tabela[medidorMundo_tabela['TotalCases'] > 10000]
medidorMundo_tabela = medidorMundo_tabela.groupby('Country/Region').sum()
medidorMundo_tabela.style.background_gradient(cmap='RdPu')
```

#### Avaliando Dados Para o Brasil

```
data_brasil = data[data['Country/Region']=='Brazil']
data_brasil = data_brasil[data_brasil.Confirmed > 0]
data_brasil.head()
```

```
fig = px.line(data_brasil,x='Date',y = 'New cases',title='Taxa de infecção', render_mode='png')
fig.show()
```

```
fig = px.line(data_brasil,x='Date',y = 'New deaths',title='Taxa de mortes',render_mode='png')
fig.show()
```

```
fig = px.line(data_brasil,x='Date',y = 'New recovered',title='Taxa de recuperação',render_mode='png')
fig.show()
```

#### Comparação de mortes por milhões de habitantes dos 10 países mais afetados

```
fig = px.bar(medidorMundo.head(10), y='Deaths/1M pop',x='Country/Region',color='WHO Region',height=400)
fig.update_layout(title='Comparação de mortes por milhões dos 10 países mais afetados',xaxis_title='Country',yaxis_title='Deaths/Million',template="plotly_dark")
fig.show(render_mode='png')
```

##### Comparação de casos por milhões de habitantes dos 10 países mais afetados

```
fig = px.bar(medidorMundo.head(10), y='Tot Cases/1M pop',x='Country/Region',color='WHO Region',height=400)
fig.update_layout(title='Comparação de casos / milhão dos 10 países mais afetados',xaxis_title='Country',yaxis_title='Cases/Million',template="plotly_dark")
fig.show(render_mode='png')
```

#### Porcentagem do total de mortes nos 20 países mais afetados

```
fig = px.pie(medidorMundo.head(20),values='TotalDeaths',names='Country/Region',title='Porcentagem do total de mortes nos 20 países mais afetados')
fig.show(render_mode='png')
```
